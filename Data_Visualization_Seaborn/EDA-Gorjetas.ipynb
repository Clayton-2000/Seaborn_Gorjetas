## Constatando se os dois casos realmente influenciam no valor da gorjeta

_Os grupos que pediram e que não pediram sobremesa se comportam de formas diferentes quando se trata do valor de gorjeta. Utilizamos os dados de uma amostra, pois não podemos verificar o comportamento de todas as pessoas do mundo e colocar os dados dentro de nosso plot._
_Porém, precisaremos pensar se os clientes que futuramente virão ao nosso restaurante se comportarão da mesma maneira ou não. Vimos antes que, adicionando a porcentagem, as diferenças entre os grupos fica bem mais visível._
_Neste passo, descobriremos se a população geral que consome a sobremesa e a que não consome se comportam igual ou diferentemente da amostra através de testes de hipótese._

## Teste de Hipótese

**H<sup>null</sup>**

>**A distribuição da taxa da gorjeta é a mesma nos dois grupos?**

**H<sup>alt</sup>**

>**A distribuição da taxa da gorjeta não é a mesma nos dois grupos?**

```python
# importando a biblioteca Ranksums a partir de scipy.stats
from scipy.stats import ranksums
```

```python
# tratando o valor da porcentagem de todas as pessoas que pediram a sobremesa a partir da montagem de uma query
sobremesa = gorjetas.query('sobremesa == "Sim"').porcentagem
```

```python
# tratando o valor da porcentagem de todas as pessoas que não pediram a sobremesa
sem_sobremesa = gorjetas.query('sobremesa == "Não"').porcentagem
```

```python
# agora que já temos os dois grupos para nossa análise
# verificaremos as hipóteses através de ranksums(), passando os dois grupos
r = ranksums(sobremesa, sem_sobremesa)
```

```python
# visualização formal de p-value
print('O valor do p-value é {}'.format(r.pvalue))
```

_O resultado '0.52' significa que, apesar de os dois grupos terem dado gorjetas de valores diferentes entre si, a diferença é insignificante do ponto de vista matemático. Ou seja, não poderemos dizer que a população geral será diferente também._
_Cientes disso, continuaremos somente com nossa hipótese nula, não aceitando mais a hipótese alternativa. Para aceitarmos a hipótese alternativa, o valor de p-value deveria ser igual ou menor do que 0.05_

**H<sup>null</sup>**

>**A distribuição da taxa da gorjeta é a mesma nos dois grupos.**

## Teste de Hipótese

**H<sup>null</sup>**

>**A distribuição do valor da conta é igual no sábado e no domingo?**

**H<sup>alt</sup>**

>**A distribuição do valor da conta não é igual no sábado e no domingo?**

```python
# montando a query() que trará o valor da conta de 'Domingo'
valor_conta_domingo = gorjetas.query('dia_da_semana == "Domingo"').valor_da_conta
```

```python
# montando a query() que trará o valor da conta de 'Sábado'
valor_conta_sabado = gorjetas.query('dia_da_semana == "Sábado"').valor_da_conta
```

```python
# rodando o ranksums() para ver se os dois grupos são iguais ou diferentes
r2 = ranksums(valor_conta_domingo, valor_conta_sabado)
print('O valor do p-value é {}'.format(r2.pvalue))
```

_O resultado de pvalue será 0.35, ou seja, apenas aceitaremos a hipótese alternativa se este número for menor do que 0.05; como é maior, apenas nossa hipótese nula é válida. Isso significa que a distribuição do valor da conta é igual no sábado e no domingo. Portanto, mesmo que o sábado seja mais frequentado e o domingo possua a maior média de gorjetas, não são diferentes no que diz respeito ao valor da conta._

**H<sup>null</sup>**

>**A distribuição do valor da conta é igual no sábado e no domingo.**

## Teste de Hipótese 2

_Faremos um teste com pvalue relacionado à porcentagem que possui uma média bastante parecida para as duas horas do dia, verificando se são grupos diferentes de fato ou não._

**H<sup>null</sup>**

>**A distribuição da taxa da gorjeta é igual no jantar e no almoço?**

**H<sup>alt</sup>**

>**A distribuição da taxa da gorjeta não é igual no jantar e no almoço?**

```python
# montando a query() que trará a distribuição da taxa da gorjeta no 'Almoço'
porcentagem_almoco = gorjetas.query('hora_do_dia == "Almoço"').porcentagem
```

```python
# montando a query() que trará a distribuição da taxa da gorjeta no 'Jantar'
porcentagem_jantar = gorjetas.query('hora_do_dia == "Jantar"').porcentagem
```

```python
# rodando o ranksums() para ver se os dois grupos são iguais ou diferentes
r4 = ranksums(porcentagem_almoco, porcentagem_jantar)
print('O valor do p-value é de {}'.format(r4.pvalue))
```
